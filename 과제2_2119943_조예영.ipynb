1. 재귀 초급 문제  
팰린드롬이란, 앞에서부터 읽었을 때와 뒤에서부터 읽었을 때가 같은 문자열을 말한다.  
어떤 문자열이 팰린드롬인지 판별하는 문제는 재귀 함수를 이용해 쉽게 해결할 수 있다. 아래 코드의 isPalindrome 함수는 주어진 문자열이 팰린드롬이면 1, 팰린드롬이 아니면 0을 반환하는 함수다.  
동민이는 위에 작성된 isPalindrome 함수를 이용하여 어떤 문자열이 팰린드롬인지 여부를 판단하려고 한다.  
문자열 S를 isPalindrome 함수의 인자로 전달하여 팰린드롬 반환값과 판별하는 과정에서 recursion 함수를 몇 번 호출하는지 아래와 같이 출력하는 프로그램을 코딩하시오.


```python
def recursion(s, l, r):
    if l >= r: return 1
    elif s[l] != s[r]: return 0
    else: return recursion(s, l+1, r-1)  

def isPalindrome(s):
    return recursion(s, 0, len(s)-1)

print('ABBA:', isPalindrome('ABBA'))
print('ABC:', isPalindrome('ABC'))
```

입력

```
5
AAA
ABBA
ABABA
ABCA
PALINDROME
```
출력

```
1 2
1 3
1 3
0 2
0 1
```



````python
def recursion(s, l, r, count):
    if l >= r: #한 글자와 홀수 문자열의 마지막은 비교 대상이 없으므로 1반환
        return 1, count
    elif s[l] != s[r]: #팬린드롬이 아니므로 0반환
        return 0, count
    else:
        count += 1 # 카운트
        return recursion(s, l + 1, r - 1, count)

def is_palindrome(s):
    count = 1 #카운트 시작
    result, count = recursion(s, 0, len(s) - 1, count)
    return result, count

# 입력
n = int(input())  # 문자열의 개수를 입력
s = []
for _ in range(n):
    x = input()  # n개의 문자열을 입력
    s.append(x) # 입력 받은 문자열을 s에 저장
for s1 in s:
    result, count = is_palindrome(s1) #팰린드롬 실행
    print(result, count) #결과 출력
````
