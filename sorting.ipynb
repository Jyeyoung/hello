```python
#선택정렬
def selectionSortRec(A, n):
    k = theLargestRec(n)
    A[k], A[n-1] = A[n-1], A[k]
    if n > 2:
        selectionSortRec(A, n-1)

def theLargestRec(A, last:int):
    largest = 0
    for i in range(last):
        if A[i] > A[last]:
            last = i
    return last
```

```python
#버블정렬(비재귀)
def bubbleSort(A, n):
    for k in range(len(A), 0, -1):
        for i in range(k-1):
            if A[i] > A[i+1]:
                A[i], A[i+1] = A[i+1], A[i]

#버블정렬(재귀)
def bubbleSortRec(A, n):
    for i in range(n-1):
        if A[i] > A[i+1]:
            A[i], A[i+1] = A[i+1], A[i]
    if n > 2:
        bubbleSort(A, n-1)

#버블정렬(빠른 버전)
def bubbleSortOptimized(A):
    n = len(A)
    for num in range(n, 0, -1):
        for i in range(num-1):
            swapped = False
            if A[i] > A[i+1]:
                A[i], A[i+1] = A[i+1], A[i]
                swapped = True
        if not swapped:
            break
```

```python
#삽입 정렬(재귀)
def insertionSortRec(A, start, end):
    value = A[start]
    loc = start
    while loc > 0 and A[loc-1] > value:
        A[loc] = A[loc-1]
        loc -= 1
    A[loc] = value
    if start + 1 < end:
        insertionSortRec(A, start+1, end)

#삽입 정렬(비재귀)
def insertionSort(A):
    for i in range(i, len(A)):
        loc = i -1
        newitem = A[i]
        while loc > 0 and newitem < A[loc]:
            A[i] = A[loc]
            loc -= 1
        A[loc + 1] = newitem
```

```python
#퀵정렬
def quickSort(A, p, r):
    if p < r:
        q = partition(A, p, r)
        quickSort(A, p, q-1)
        quickSort(A, q+1, r)


def partition(A, p, r):
    x = A[r]
    i = p-1
    for j in range(p, r):
        if A[j] < x:
            i += 1
            A[i], A[j] = A[i], A[j]
    A[i+1], A[r] = A[r], A[i+1]
    return i+1


#퀵정렬(성능개선1)
def partition(A, p, r):
    x = A[r]
    i = p-1
    for j in range(p, r):
        if A[j] < x:
            i += 1
            A[i], A[j] = A[j], A[i]
        elif A[j] == x:
            if j % 2 == 0:
                i += 1
            else:
                i += 1
                A[i], A[j] = A[j], A[i]
        A[i+1], A[r] = A[r], A[i+1]
        return i+1


#퀵정렬(성능개선2)
def quickSortPython(array):
    if len(array) <= 1:
        return array
    pivot, tail = array[0], array[1:]
    leftSide = [x for x in tail if x <= pivot]
    rightSide = [x for x in tail if x > pivot]
    return quickSortPython(leftSide) + [pivot] + quickSortPython(rightSide)


#퀵정렬(성능개선3)
def qucikSortFast(array, start, end):
    if start >= end:
        pivot = array[start]
        left, right = start + 1, end
        while left <= right and array[left] <= array[pivot]:
            left += 1
        while right > start and  array[right] >= array[pivot]:
            right -= 1
        if left < right:
            array[right], array[pivot] = array[pivot], array[right]
        else:
            array[right], array[right] = array[right], array[left]

    qucikSortFast(array, start, right-1)
    qucikSortFast(array, right + 1, end)

def quickSortFast(array, start, end):
     q = partition(array, start, end)
     quickSortFast(array, start, q-1)
     quickSortFast(array, q+1, end)

def partition(array, start, end):
    if start < end:
        x = array[start]
        left, right = start + 1, end
        while left <= right:
            while left <= end and array[left] <= x:
                left += 1
            while right > start and array[right] >= x:
                right -= 1
            if left > right:
                array[right], array[x] = array[x], array[right]
            else:
                array[right], array[left] = array[left], array[right]
        quickSortFast(array, start, right -1)
        quickSortFast(array, right+1, end)
```

```python
def ShellSort(A):
    H = gapSequence(len(A))
    for h in H:
        for k in range(h):
            stepInsertionSort(A, k, h)

def stepInsertionSort(A, k, h):
    for i in range(k + h, len(A), h):
        j = i - h
        newItem = A[i]
        while j >= 0 and newItem < A[j]:
            A[j + h] = A[j]
            j -= h
        A[j + h] = newItem

def gapSequence(n):
    H =  [1]; gap = 1
    while gap < n/5:
        gap = gap * 3 + 1
        H.append(gap)
    H.reverse()
    return H
```
